```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

Line Chart

```python
line = df.groupby('age').mean(numeric_only=True).reset_index()
plt.figure(figsize=(14, 6))
# line chart
sns.lineplot(x='age',y='SysBP',data=line)
```

Bar plot

```python
plt.figure(figsize=(14, 6)) 
sns.barplot(x='age',y='HR',hue='sex',data=df)
```

Pie Chart

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

Box Plot


```python
sns.boxplot(df['SysBP'])
#this shows count of male and females were identicals
```

Histogram

```python
plt.figure(figsize=(20, 6))
plt.hist(df['DiaBP'])
#the bp of people lies in the median
```

Scatter plot

```python
plt.figure(figsize=(14, 6))
sns.scatterplot(x='age',y='BMI',hue='sex',data=df)
plt.show()
#
```
